```python
import pandas as pd
import numpy as np
df=pd.read_csv('rs_dataframe.csv')
pd.set_option('display.max_columns', None)
```

```python
df=df[df['tradeTypeName']!='매매']
df=df.reset_index(drop=True)
```

```python
df.info()
```

```python
df['2nPrc_str']=df['rentPrc']+'*24+'+df['dealOrWarrantPrc']
nPrc=[]
for i in df['2nPrc_str']:
    try :
        i=i.replace(',','')
        i=i.replace('억','0000')
        i=i.replace(' ','+')
        i=eval(i)*10000
        nPrc.append(int(i))
    except:
        nPrc.append(i)
        continue
df['2nPrc']=nPrc
```

```python
df
```

```python
df['평당가격']=df['2nPrc']/df['area2']
df['가성비']=df['flow_pop']/df['평당가격']
df=df.sort_values('가성비',ascending=False)
```

```python
Rank=[]
for i in range(1,len(df)+1):    
    Rank.append(i)
df['가성비_Rank']=Rank
df
```

```python
df=df.reset_index(drop=True)
```

```python
df.drop(['Unnamed: 0'],inplace=True,axis=1)
#df.drop(['Unnamed: 0.1'],inplace=True,axis=1)
```

```python
df.columns
```

```python
floor_a = []
floor_b = []
for i in df.floorInfo.values:
    floor_a.append(i.split("/")[0])
    floor_b.append(i.split("/")[1])
```

```python
df['floor_estate'] = floor_a
df['floor_buliding'] = floor_b
```

```python
df.to_csv('rs_dataframe.csv',index=False)
```
